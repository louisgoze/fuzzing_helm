```python
def find_type(names, definitions, api_type):
    current_definition = api_type
    for name in names:
        properties = definitions[current_definition]['properties']
        if name in properties:
            property_details = properties[name]
            property_type = property_details.get('type')
            if not property_type:
                ref = property_details['$ref']
                current_definition = ref.split('/')[-1]
            else:
                return property_type
        else:
            return "Property not found"
    return property_type

result = find_type(["metadata","namespace"], definitions, 'io.k8s.api.apps.v1.Deployment')
print(result)
```

```python
import random
import yaml
import csv
import os
import json
import logging
import subprocess

log: logging.Logger = logging.getLogger('custom_logger')

def _execute_command_with_error(command: str):
    process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    output, error = process.communicate()
    try:
        return output.decode("utf-8"), error.decode("utf-8") if error else None
    except UnicodeDecodeError:
        return output.decode("latin-1"), error.decode("latin-1") if error else None

def compare_checkov(before_output, after_output) -> tuple[int, int]:
    """Returns the amount of failed tests before and after"""
    before = json.loads(before_output)
    after = json.loads(after_output)
    return before["summary"]["failed"], after["summary"]["failed"]

def compare_kubesec(before_output, after_output) -> tuple[int, int]:
    """Retourne le nombre de tests échoués avant et après"""
    before = json.loads(before_output)
    after = json.loads(after_output)
    try:
        return len(before[0]["scoring"]["advise"]), len(after[0]["scoring"]["advise"])
    except:
        return len(before[0]["scoring"]["advise"]), 'Error'

def compare_kubeaudit(before_output, after_output) -> tuple[int, int]:
    """Retourne le nombre de tests échoués avant et après"""
    before_lines = before_output.strip().split("\n")
    after_lines = after_output.strip().split("\n")
    return len(before_lines), len(after_lines)

def compare_kubescore(before_output, after_output) -> tuple[int, int]:
    """Retourne le nombre de tests échoués avant et après"""
    before = before_output.split('[')
    after = after_output.split('[')
    return len(before), len(after)

# Charger le fichier CSV contenant les variables et leurs valeurs possibles
csv_input_path = 'output/variables_with_possible_values.csv'
variables = []

with open(csv_input_path, mode='r') as infile:
    reader = csv.DictReader(infile)
    for row in reader:
        variables.append(row)

# Charger le fichier YAML d'origine
yaml_file_path = 'input/example.yml'
with open(yaml_file_path, 'r') as file:
    yaml_data = yaml.safe_load(file)

# Dépendances entre les variables
dependencies = {
    'metadata.labels.app': ['spec.selector.matchLabels.app', 'spec.template.metadata.labels.app'],
    # Ajoutez d'autres dépendances ici
}

# Fonction pour définir la valeur à la clé donnée dans un dictionnaire imbriqué
def set_nested_key(data, keys, value):
    for key in keys[:-1]:
        if isinstance(data, list):
            data = data[int(key)]
        else:
            data = data.setdefault(key, {})
    if isinstance(data, list):
        data[int(keys[-1])] = value
    else:
        data[keys[-1]] = value

# Fonction pour mettre à jour une variable dans le fichier YAML
def update_variable(yaml_data, variable, value, dependencies):
    variable_path = variable['Variable']
    
    # Convertir le chemin de la variable en une liste de clés
    keys = variable_path.replace('[', '.').replace(']', '').split('.')
    keys = [int(k) if k.isdigit() else k for k in keys]
    
    # Mettre à jour la valeur dans le dictionnaire YAML
    set_nested_key(yaml_data, keys, value)
    
    # Mettre à jour les variables dépendantes
    if variable_path in dependencies:
        for dependent_variable in dependencies[variable_path]:
            dependent_keys = dependent_variable.replace('[', '.').replace(']', '').split('.')
            dependent_keys = [int(k) if k.isdigit() else k for k in dependent_keys]
            set_nested_key(yaml_data, dependent_keys, value)
    
    return yaml_data, variable_path, value

# Fonction pour nettoyer les noms de fichiers
def clean_filename(value):
    return value.replace('/', '_').replace('@', '_').replace(':', '_').replace(' ', '_')

# Créer le répertoire de sortie s'il n'existe pas
output_directory = 'v2output/'
os.makedirs(output_directory, exist_ok=True)

# Créer le répertoire de résultats s'il n'existe pas
results_directory = 'results/'
os.makedirs(results_directory, exist_ok=True)

# Générer tous les fichiers YAML fuzzés
fuzzed_files = []
for variable in variables:
    possible_values = variable['Possible Values'].split(', ')
    for value in possible_values:
        # Charger à nouveau le fichier YAML d'origine pour chaque modification
        with open(yaml_file_path, 'r') as file:
            yaml_data = yaml.safe_load(file)
        
        updated_yaml_data, updated_variable, new_value = update_variable(yaml_data, variable, value, dependencies)
        
        # Sauvegarder le fichier YAML mis à jour
        variable_name = updated_variable.replace('.', '_').replace('[', '_').replace(']', '')
        cleaned_value = clean_filename(new_value)
        updated_yaml_file_path = os.path.join(output_directory, f'output_fuzz_{variable_name}_{cleaned_value}.yml')
        with open(updated_yaml_file_path, 'w') as file:
            yaml.safe_dump(updated_yaml_data, file)
        fuzzed_files.append(updated_yaml_file_path)

print("Tous les fichiers YAML mis à jour ont été sauvegardés dans le répertoire 'v2output'.")

# Chemin du fichier YAML de base
base_yaml_path = 'input/example.yml'

# Comparer les résultats de chaque fichier fuzzé avec le fichier de base
for fuzzed_file in fuzzed_files:
    result = {"fuzzed_file": fuzzed_file}
    print(f"Comparaison des résultats pour {fuzzed_file} :")
    
    # Comparaison avec Checkov
    base_checkov_output, _ = _execute_command_with_error(f"checkov -f {base_yaml_path} --output json")
    fuzzed_checkov_output, _ = _execute_command_with_error(f"checkov -f {fuzzed_file} --output json")
    base_failed, fuzzed_failed = compare_checkov(base_checkov_output, fuzzed_checkov_output)
    result["checkov"] = {"base_failed": base_failed, "fuzzed_failed": fuzzed_failed}
    print(f"Checkov - Tests échoués avant: {base_failed}, après: {fuzzed_failed}")
    
    # Comparaison avec Kubesec
    base_kubesec_output, _ = _execute_command_with_error(f"kubesec scan {base_yaml_path} --format json")
    fuzzed_kubesec_output, _ = _execute_command_with_error(f"kubesec scan {fuzzed_file} --format json")
    base_failed, fuzzed_failed = compare_kubesec(base_kubesec_output, fuzzed_kubesec_output)
    result["kubesec"] = {"base_failed": base_failed, "fuzzed_failed": fuzzed_failed}
    print(f"Kubesec - Tests échoués avant: {base_failed}, après: {fuzzed_failed}")
    
    # Comparaison avec Kubeaudit
    base_kubeaudit_output, _ = _execute_command_with_error(f"kubeaudit all -f {base_yaml_path}")
    fuzzed_kubeaudit_output, _ = _execute_command_with_error(f"kubeaudit all -f {fuzzed_file}")
    base_failed, fuzzed_failed = compare_kubeaudit(base_kubeaudit_output, fuzzed_kubeaudit_output)
    result["kubeaudit"] = {"base_failed": base_failed, "fuzzed_failed": fuzzed_failed}
    print(f"Kubeaudit - Tests échoués avant: {base_failed}, après: {fuzzed_failed}")
    
    # Comparaison avec Kubescore
    base_kubescore_output, _ = _execute_command_with_error(f"kube-score score {base_yaml_path}")
    fuzzed_kubescore_output, _ = _execute_command_with_error(f"kube-score score {fuzzed_file}")
    base_failed, fuzzed_failed = compare_kubescore(base_kubescore_output, fuzzed_kubescore_output)
    result["kubescore"] = {"base_failed": base_failed, "fuzzed_failed": fuzzed_failed}
    print(f"Kubescore - Tests échoués avant: {base_failed}, après: {fuzzed_failed}")

    # Enregistrer les résultats dans un fichier JSON
    cleaned_value = clean_filename(fuzzed_file)
    result_file_path = os.path.join(results_directory, f'results_{cleaned_value}.json')
    with open(result_file_path, 'w') as result_file:
        json.dump(result, result_file, indent=4)

print("Tous les résultats de comparaison ont été sauvegardés dans le répertoire 'results'.")
```
